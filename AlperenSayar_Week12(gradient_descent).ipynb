```python
# Soru 1) Gradient Descent algortimasını kullanarak x^2+2*x+1 fonksiyonu için gloabil minumum noktasını bulunuz.
# Soru 2) Learning rate parametresini 0.001, 0.01,0.0001,0.1 olarak deneyerek sonuçlarını karşılaştırınız.
```

```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import sklearn.datasets as dt
```

```python
import numpy as np

def gradient_descent(start, gradient, learn_rate, max_iter, tol=1e-3):
 
  x = start
  for _ in range(max_iter):
    diff = learn_rate*gradient(x)
    if np.abs(diff)<=tol:
      break    
    x = x - diff
  
  return x
```

```python
def func(x):
  return x**2+2*x+1



def func2(x):
  return x+1
```

```python
learning=[0.001, 0.01,0.0001,0.1]
```

```python
for i in range(len(learning)):
  print("For learning rate:",learning[i],"->",gradient_descent(-3, func2, learning[i], 1000))
```

```python

linspace = np.linspace(-10,10,100)
plt.figure(figsize=(15,15))
plt.plot(linspace,func(linspace))
plt.title(" Func")
plt.xlabel("X")
plt.ylabel("y")
plt.show()
```
